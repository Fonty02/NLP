```python
#Function1
def freq(w,txt ):
        freqDist = nltk.FreqDist(txt)
        return freqDist.freq(w)

print(freq("whale", text1))
```

```python
def cfreq(wlist,txt):
    #compute the comulative frequency of a list of words in text txt
    freqDist = nltk.FreqDist(txt)
    tot=0
    for w in wlist:
        tot+=freqDist.freq(w)
    return tot
print(cfreq(["monstrous", "very"], text1))
```

```python
def cfreq_n(txt,n):
    freqDist = nltk.FreqDist(txt)
    total=0
    mostCommon = freqDist.most_common(n)
    for w in mostCommon:
        total+=freqDist.freq(w[0])
    return total

print(cfreq_n(text1, 10))
```

```python
def cfreq_length(txt,k):
    freqDist = nltk.FreqDist(txt)
    total=0
    for word in freqDist:
        if len(word) >= k:
            total+=freqDist.freq(word)
    return total

print(cfreq_length(text1, 5))
```
