```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text)
    english_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - english_vocab  # words not in the English vocabulary
    return sorted(unusual)
print(unusual_words(gutenberg.words('austen-emma.txt')))
print(unusual_words(gutenberg.words('austen-persuasion.txt')))

from nltk.corpus import stopwords
print(stopwords.words('english'))
```

### Ex1 -> Define a conditional frequency distribution over the Names corpus that allows you to see which initial letters are more frequent for males vs. females ###

```python
from nltk.corpus import names

print(names.fileids()) #female, male
# Get the names from female.txt and male.txt
female_names = names.words('female.txt')
male_names = names.words('male.txt')

cfd = nltk.ConditionalFreqDist(
    (gender, name[0])  # (condition, sample) pairs: (gender, first letter)
    for gender in ['female', 'male']
    for name in (female_names if gender == 'female' else male_names)
)

cfd.tabulate()
```

Define a function supergloss(s) that takes a
synset s as its argument and returns a string
consisting of the concatenation of the
definition of s, and the definitions of all the
hypernyms and hyponyms of s


```python
def supergloss(s):
    type_of_s = s.hyponyms()
    s_type_of = s.hypernyms()
    gloss = s.definition()
    hyponym_glosses = [h.definition() for h in type_of_s]
    hypernym_glosses = [h.definition() for h in s_type_of]
    result = gloss + " " + " ".join(hyponym_glosses) + " " + " ".join(hypernym_glosses)
    return result

print(supergloss(wn.synsets('man')[0]))  #example
```

Write a function word_freq() that takes a word
and the name of a section of the Brown Corpus
as arguments, and computes the frequency of
the word in that section of the corpus


```python
def word_freq(word,sectionOfBrown):
    files=brown.fileids([sectionOfBrown])
    count=0
    for f in files:
        w=brown.words(fileids=f)
        
        fd=nltk.FreqDist(a.lower() for a in w)
        count+=fd[word.lower()]
        
    return count
    

    # Test the word_freq function on various sections of the Brown corpus
print("Frequency of 'of' in news:", word_freq('of', 'news'))
```

The polysemy of a word is the number of senses
it has. Compute the average polysemy of nouns,
verbs, adjectives and adverbs according to
WordNet

```python
def average_polysemy(pos):
    polysemy = []
    for synset in wn.all_synsets(pos):
        polysemy.append(len(synset.lemma_names()))
    return sum(polysemy) / len(polysemy)
print("Average polysemy of nouns:", average_polysemy('n'))
print("Average polysemy of verbs:", average_polysemy('v'))
print("Average polysemy of adjectives:", average_polysemy('a'))
print("Average polysemy of adverbs:", average_polysemy('r'))
```

Use one of the predefined similarity measures to
score the similarity of each of the following pairs of
words. Rank the pairs in order of decreasing
similarity:


```python
from nltk.corpus import wordnet as wn

# List of word pairs to compare
word_pairs = [
    ('car', 'automobile'), ('gem', 'jewel'), ('journey', 'voyage'), ('boy', 'lad'),
    ('coast', 'shore'), ('asylum', 'madhouse'), ('magician', 'wizard'), ('midday', 'noon'),
    ('furnace', 'stove'), ('food', 'fruit'), ('bird', 'cock'), ('bird', 'crane'),
    ('tool', 'implement'), ('brother', 'monk'), ('lad', 'brother'), ('crane', 'implement'),
    ('journey', 'car'), ('monk', 'oracle'), ('cemetery', 'woodland'), ('food', 'rooster'),
    ('coast', 'hill'), ('forest', 'graveyard'), ('shore', 'woodland'), ('monk', 'slave'),
    ('coast', 'forest'), ('lad', 'wizard'), ('chord', 'smile'), ('glass', 'magician'),
    ('rooster', 'voyage'), ('noon', 'string')
]

# Function to calculate similarity between two words
def word_similarity(pair):
    word1, word2 = pair
    similarity=0
    synsets1 = wn.synsets(word1)
    synsets2 = wn.synsets(word2)
    for s1 in synsets1:
        for s2 in synsets2:
            similarity=max(similarity,wn.path_similarity(s1,s2))
    return similarity

l=[]
for pair in word_pairs:
    l.append(word_similarity(pair))

    # Combine word pairs with their similarity scores
    similarity_pairs = list(zip(word_pairs, l))

    # Sort by similarity score in descending order
    sorted_pairs = sorted(similarity_pairs, key=lambda x: x[1], reverse=True)

    # Print the ranked pairs
    print("Word pairs ranked by similarity (highest to lowest):")
    for i, ((word1, word2), similarity) in enumerate(sorted_pairs, 1):
        print(f"{i}. {word1:<10} - {word2:<10} : {similarity:.4f}")
```
