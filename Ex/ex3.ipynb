```python
print ("*** some examples of usage of the method re.search ***")
wordlist = [w for w in nltk.corpus.words.words('en') if w.islower()]  #this returns a list of all words in the corpus that are lowercase
matchedtokens1 = [w for w in wordlist if re.search('ed$', w)] #this returns a list of all words in the corpus that end with 'ed'
matchedtokens2 = [w for w in wordlist if re.search('^..j..t..$', w)] #this returns a list of all words in the corpus that have 'j' and 't' as the third and fifth letters respectively
matchedtokens3 = [w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)] #this returns a list of all words in the corpus that have 'g', 'h' or 'i' as the first letter, 'm', 'n' or 'o' as the second letter, 'j', 'l' or 'k' as the third letter and 'd', 'e' or 'f' as the fourth letter
print (matchedtokens1)
print (matchedtokens2)
print (matchedtokens3)

#for regular expression, ^ means beginning



word = 'supercalifragilisticexpialidocious'
vowels=re.findall(r'[aeiou]', word)
print(vowels)
print(len(vowels))


matchedtokens4 = [w for w in word if re.search(r'[aeiou]', w)]
print(matchedtokens4)


#findall Cerca tutte le occorrenze del pattern nella stringa. Restituisce una lista con tutte le corrispondenze trovate
#search Cerca la prima occorrenza del pattern nella stringa. Restituisce un oggetto Match, se trova una corrispondenza.
```

```python
### findall enables us to search for regex patterns over tokens
### only matched word
moby = nltk.Text(gutenberg.words('melville-moby_dick.txt'))
print (moby.findall(r"<a> (<.*>) <man>"))
# three-word phrases ending with the word bro
from nltk.corpus import nps_chat
chat = nltk.Text(nps_chat.words())
seq2=chat.findall(r"<.*> <.*> <bro>")
print (seq2)
### sequences of three or more words starting with the letter l [3]
seq3=chat.findall(r"<l.*> {3,}*")  #{3,} sigifica dividi in gruppo da 3
print (seq3)
```

```python
### finding expressions of the form x and other ys allows us to discover hypernyms
hobbies_learned = nltk.Text(brown.words(categories=['hobbies',
'learned']))
patterns=hobbies_learned.findall(r"<\w*> <and> <other> <\w*s>")
print (patterns)
```

```python
raw = """'When I'M a Duchess,' she said to herself, (not in a very hopeful tone
... though), 'I won't have any pepper in my kitchen AT ALL. Soup does very
... well without--Maybe it's always pepper that makes people hottempered,'..."""
tokens=raw.split()
print ("tokens - simple split:", tokens)
### same result with a regex
### splitting on any whitespace
tokens=re.split(r'\s+',raw)
#so uguali

print ("tokens - simple split:", tokens)
### with \W+ we match all characters other than letters, digits or underscore.
### We can use \W in a simple regular expression to split the input on anything other than a word character
tokens=re.split(r'\W+', raw)
print ("tokens - \W+ split:", tokens)


#quindi praticamente \W+ fa lo stesso lavoro di nltk.Text() ma non gestisce punteggiatura, abbreviazioni, etc...
```

```python
### matching the words instead of the spaces.
tokens=re.findall(r'\w+|\S\w*', raw)
print ("tokens - findall:", tokens)
```

```python
### NLTK regex tokenizer
text = 'That U.S.A. poster-print costs $12.40...'
pattern = r'''(?x) # set flag to allow verbose regexps
(?:[A-Z]\.)+ # abbreviations, e.g. U.S.A.
| \w+(?:-\w+)* # words with optional internal hyphens
| \$?\d+(?:\.\d+)?%? # currency and percentages, e.g. $12.40, 82%
| \.\.\. # ellipsis
| [][.,;"'?():-_`] # these are separate tokens; includes ], [
'''
tokens=nltk.regexp_tokenize(text, pattern)
print ("tokens - NLTK regex tokenizer:", tokens)
```

```python
raw = """DENNIS: Listen, strange women lying in ponds distributing swords
is no basis for a system of government. Supreme executive power derives from
a mandate from the masses, not from some farcical aquatic ceremony."""
tokens = word_tokenize(raw)
print ("Raw Text", raw)
print ("TOKENS:", tokens)
porter = nltk.PorterStemmer()
lancaster = nltk.LancasterStemmer()
porstems=[porter.stem(t) for t in tokens]
lanstems=[lancaster.stem(t) for t in tokens]
print ("Porter stems:", porstems)
print ("Lancaster stems:", lanstems)
```

```python
wnl = nltk.WordNetLemmatizer()
wnlemmas=[wnl.lemmatize(t) for t in tokens]
print ("WordNet Lemmas:", wnlemmas)
print ("Lemmatizing 'arguing' with WN lemmatizer:")
print(wnl.lemmatize("arguing", pos=u'a'))
print(wnl.lemmatize("arguing", pos=u'v'))
print ("Lemmatizing 'lying' with WN lemmatizer:")
print(wnl.lemmatize("lying", pos=u'v'))  #pos is the part of speech tag

#stemming vs lemmatization
# stemming: rimuove affissi e ridurre la parola alla sua radice
# lemmatization: riduce la parola alla sua forma base (lemma) e richiede un dizionario per trovare la forma base
```

Load the text obama2013.txt

Split into sentences (each sentence is now an individual document)

Create a TF-IDF model 

```python
import numpy as np
import math
final=[]
porter = nltk.PorterStemmer()
sentences=inaugural.sents(fileids=['2013-Obama.txt'])
for i in range(len(sentences)):
    tokens = list(porter.stem(w.lower()) for w in sentences [i] if re.search('[A-Za-z0-9]', w))
    final.append(tokens)

 #final lista di liste, ogni lista  una sentence (quindi documento)

vocabulary=set()
for f in final:
    for t in f:
        vocabulary.add(t)
vocabulary=list(vocabulary)


print(vocabulary)
tf_idf_matrix=np.zeros(shape=(len(vocabulary),len(final)))
for i in range(len(vocabulary)):
    for j in range(len(final)):
        tf_idf_matrix[i][j]=final[j].count(vocabulary[i])


for i in range(len(vocabulary)):
    term=vocabulary[i]
    doc_count = sum(1 for doc in final if term in doc)
    idf = math.log(len(final)) if doc_count>0 else 0
    tf_idf_matrix[i]*=idf

scores_value=[]
for j in range(len(final)):
    scores_value.append(np.average(tf_idf_matrix[:,j]))

relevance=[j/sum(scores_value) for j in scores_value]
print(relevance)

relevant=[]

for k in range(len(final)):
    if relevance[k]>np.average(relevance):
        relevant.append(final[k])

print(relevant)







    







```
